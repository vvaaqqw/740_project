# plot predictions of BTC prices VS real BTC prices

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ann_data = pd.read_csv('ANN/btc_NN.csv')
lstm_data = pd.read_csv('LSTM/btc_NN.csv')
gbrt_data = pd.read_csv('GBRT/btc_gbrt.csv')
rf_data = pd.read_csv('RF/btc_rf.csv')
```

```python
x= pd.to_datetime(ann_data['Date'])
real_y = ann_data['Y_test']
ann_y = ann_data['y_pred']
lstm_y = lstm_data['y_pred']
gbrt_y = gbrt_data['y_pred']
rf_y = rf_data['y_pred']
```

```python
labels=['Actual BTC','ANN 30-days','LSTM 30-days','GBRT 30-days','Random Forest 30-days']
plt.figure(figsize=(13,8))
plt.xlabel('Date')
plt.ylabel('BTC price in USD')
plt.plot(x, real_y, '.-')
plt.plot(x, ann_y, '--')
plt.plot(x, lstm_y)
plt.plot(x, gbrt_y, '--')
plt.plot(x, rf_y, '--')
plt.legend(labels)
```

```python
lstm_eth_data = pd.read_csv('LSTM/eth_NN.csv')
ann_eth_data = pd.read_csv('ANN/eth_NN.csv')

lstm_ltc_data = pd.read_csv('LSTM/ltc_NN.csv')
ann_ltc_data = pd.read_csv('ANN/ltc_NN.csv')

ann_eth_y = ann_eth_data['y_pred']
ann_ltc_y = ann_ltc_data['y_pred']

lstm_eth_y = lstm_eth_data['y_pred']
lstm_ltc_y = lstm_ltc_data['y_pred']
```

```python
labels=['ANN_LTC', 'LSTM_LTC']
plt.figure(figsize=(13,8))
plt.xlabel('Date')
plt.ylabel('Pridicted LTC price in USD')
plt.plot(x, ann_ltc_y)
plt.plot(x, lstm_ltc_y)
plt.legend(labels)
```
